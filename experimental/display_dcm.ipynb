```python
from pathlib import Path

import matplotlib.pyplot as plt
from pydicom import dcmread

DATA_DIR = Path("../data")
```

```python
def display_dcm(path: Path):
    ds = dcmread(path)
    plt.imshow(ds.pixel_array, cmap="gray")


img_path = DATA_DIR / "anon_AD01_5.dcm"
display_dcm(img_path)
```
